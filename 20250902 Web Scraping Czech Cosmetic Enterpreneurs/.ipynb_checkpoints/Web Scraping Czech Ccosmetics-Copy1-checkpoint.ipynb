```python
import folium

def shift_east(lat, lon, distance_km):
    km_per_degree_lon = 71
    delta_lon = distance_km / km_per_degree_lon
    return lat, lon + delta_lon

def shift_south(lat, lon, distance_km):
    km_per_degree_lat = 111
    delta_lat = distance_km / km_per_degree_lat
    return lat - delta_lat, lon

def add_circle(map_obj, center, radius_m, color, popup_text):
    folium.Circle(
        location=center,
        radius=radius_m,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.2,
        popup=popup_text
    ).add_to(map_obj)

# Parameters
initial_lat, initial_lon = 50.896450, 12.250195
circle_radius_m = 25000
distance_km = 50
rows = 6
cols = 10
circle_color = 'blue'

# Create map
map_czech = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Step 1: Store valid circle centers
circle_centers = []

for row in range(rows):
    row_lat, row_lon = shift_south(initial_lat, initial_lon, distance_km * row)
    for col in range(cols):
        row_index = row + 1  # 1-based indexing
        col_index = col + 1

        # Skip conditions
        if (row_index == 1 and col_index in [1, 2, 7, 8, 9, 10]) or \
           (row_index == 2 and col_index == 10) or \
           (row_index == 4 and col_index in [1]) or \
           (row_index == 5 and col_index in [1, 10]) or \
           (row_index == 6 and col_index in [1, 2, 9, 10]):
            continue

        lat, lon = shift_east(row_lat, row_lon, distance_km * col)
        circle_centers.append((lat, lon))

# Step 2: Draw circles from stored positions
for i, center in enumerate(circle_centers):
    add_circle(map_czech, center, circle_radius_m, circle_color, f'Circle {i+1}')

# Display map
map_czech
```

```python
# Step 3: Organize valid circle positions by grid index
circle_grid = {}
index = 0
for row in range(rows):
    row_lat, row_lon = shift_south(initial_lat, initial_lon, distance_km * row)
    for col in range(cols):
        row_index = row + 1
        col_index = col + 1

        if (row_index == 1 and col_index in [1, 8, 9, 10]) or \
           (row_index == 2 and col_index == 10) or \
           (row_index == 4 and col_index in [1]) or \
           (row_index == 5 and col_index in [1]) or \
           (row_index == 6 and col_index in [1, 2, 10]):
            continue

        lat, lon = shift_east(row_lat, row_lon, distance_km * col)
        circle_grid[(row, col)] = (lat, lon)

# Step 4A: Calculate and store gap centers
secondary_circle_centers = []

for row in range(rows - 1):
    for col in range(cols - 1):
        try:
            c1 = circle_grid[(row, col)]
            c2 = circle_grid[(row, col + 1)]
            c3 = circle_grid[(row + 1, col)]
            c4 = circle_grid[(row + 1, col + 1)]

            gap_lat = (c1[0] + c2[0] + c3[0] + c4[0]) / 4
            gap_lon = (c1[1] + c2[1] + c3[1] + c4[1]) / 4
            secondary_circle_centers.append((gap_lat, gap_lon))
        except KeyError:
            continue  # Skip if any of the four corners are missing

# Step 4B: Add gap circles to the map
for i, center in enumerate(secondary_circle_centers):
    add_circle(map_czech, center, circle_radius_m, 'red', f'Gap Filler {i+1}')

map_czech
```

```python
import folium

# Combine both primary and secondary circle centers
all_circle_centers = circle_centers + secondary_circle_centers

# Create a new map centered over the Czech Republic
map_combined = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Add all circles to the new map
for i, center in enumerate(all_circle_centers):
    lat, lon = center
    popup_text = f'Unified Circle {i+1} {lat:.6f}, {lon:.6f}'
    add_circle(map_combined, center, circle_radius_m, 'green', popup_text)

map_combined.save("czech_circles_map.html")

# Display the new map
map_combined
```

```python
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.alert import Alert
from selenium.common.exceptions import NoAlertPresentException, UnexpectedAlertPresentException

def scrape_circle_data(base_url, business_code, radius_km, circle_data, driver):
    """
    Scrapes data for each circle center and saves HTML content to corresponding filenames.

    Parameters:
        base_url (str): Base URL for the query.
        business_code (str): Business code used in the query.
        radius_km (int or float): Radius around each center in kilometers.
        circle_data (list): List of dicts with 'center' (lat, lon) and 'filename'.
        driver (selenium.webdriver): Selenium WebDriver instance.
    """
    for index, item in enumerate(circle_data):
        lat, lon = item["center"]
        filename = item["filename"]

        # Print progress info
        print(f"Processing {index + 1} of {len(circle_data)}: {filename}")
        
        params = {
            "roleSubjektu": "P",
            "predmet": business_code,
            "kodPredmetu": business_code,
            "center": f"{lat},{lon}",
            "distance": str(radius_km)
        }

        query_string = ";".join([f"{k}={v}" for k, v in params.items()])
        full_url = f"{base_url};{query_string}"
        print(f"Accessing: {full_url}")

        driver.get(full_url)
        time.sleep(5)

        # Handle unexpected alerts
        try:
            Alert(driver).dismiss()
        except NoAlertPresentException:
            pass

        try:
            soup = BeautifulSoup(driver.page_source, "html.parser")
            with open(filename, "w", encoding="utf-8") as f:
                f.write(soup.prettify())
        except UnexpectedAlertPresentException:
            print(f"Alert blocked access for {filename}. Skipping.")
            continue

    driver.quit()
```

```python
import os
import pandas as pd
from bs4 import BeautifulSoup

def parse_entity_records(data_dir):
    records = []

    for filename in os.listdir(data_dir):
        #if filename.startswith('center') and filename.endswith('.txt'):
            with open(os.path.join(data_dir, filename), 'r', encoding='utf-8') as file:
                soup = BeautifulSoup(file.read(), 'html.parser')

                for card in soup.select('div.card.sheet'):
                    record = {'source_file': filename}

                    # Extract name
                    name_tag = card.select_one('h5')
                    record['name'] = name_tag.get_text(strip=True) if name_tag else None

                    # Extract registered_office and distance
                    address_dd = card.select_one("div[title='Corresponding entity address'] dd")
                    if address_dd:
                        span = address_dd.find("span", title="Distance from the location")
                        distance = span.get_text(strip=True) if span else None
                        full_address = address_dd.get_text(strip=True)
                        if span:
                            full_address = full_address.replace(distance, '').strip()
                        record['registered_office'] = full_address
                        record['distance_from_the_location'] = distance
                    else:
                        record['registered_office'] = None
                        record['distance_from_the_location'] = None

                    # Initialize fields
                    record['legal_type'] = None
                    record['role'] = None
                    record['RN'] = None

                    # Build label-value map from all <dt>/<dd> pairs
                    label_map = {}
                    for dl in card.select('dl.compact.tree'):
                        new_dt = soup.new_tag('dt')
                        new_dt.string = 'Legal type'
                        first_dt = dl.find('dt')
                        if first_dt:
                            first_dt.insert_before(new_dt)

                        dt_tags = dl.select('dt')
                        dd_tags = dl.select('dd')

                        for dt, dd in zip(dt_tags, dd_tags):
                            label = dt.get_text(strip=True)
                            value = dd.get_text(strip=True)
                            label_map[label] = value

                    # Assign values from label_map
                    record['legal_type'] = label_map.get('Legal type')
                    record['role'] = label_map.get('Role')
                    record['RN'] = label_map.get('RN')

                    records.append(record)

    return pd.DataFrame(records)
```

```python
df_from_primary_circles = parse_entity_records('Original center files')
df_from_primary_circles.head()
```

```python
# Step 1: Convert 'distance_from_the_location' to numeric (in km)
df_from_primary_circles["distance_km"] = df_from_primary_circles["distance_from_the_location"].str.extract(r"([\d\.]+)").astype(float)

# Step 2: Group by 'source_file' and compute aggregations
grouped = df_from_primary_circles.groupby("source_file").agg(
    count=("name", "count"),
    min_distance_km=("distance_km", "min"),
    max_distance_km=("distance_km", "max")
).reset_index()

# Step 3: Preview the result
print(grouped[grouped['count'] >= 50])
```

```python
import folium
import re

# Step 1: Filter the grouped DataFrame
filtered = grouped[grouped['count'] >= 50]

# Step 2: Extract coordinates from filenames
def extract_coords(filename):
    match = re.search(r'center_([\d\.]+)_([\d\.]+)\.txt', filename)
    if match:
        lat = float(match.group(1))
        lon = float(match.group(2))
        return lat, lon
    return None

# Step 3: Create a folium map centered over the Czech Republic
map_selected = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Step 4: Add circles for each selected center
for _, row in filtered.iterrows():
    coords = extract_coords(row['source_file'])
    if coords:
        folium.Circle(
            location=coords,
            radius=25000,  # 25 km radius
            color='red',
            fill=True,
            fill_color='red',
            fill_opacity=0.4,
            popup=f"{row['source_file']}<br>Count: {row['count']}<br>Min: {row['min_distance_km']} km<br>Max: {row['max_distance_km']} km"
        ).add_to(map_selected)

# Step 5: Display or save the map
map_selected
```

```python
import folium
import math
import re

# Conversion constants
km_per_deg_lat = 111
km_per_deg_lon = 71
small_radius_km = 5
large_radius_km = 25

# Hex grid spacing (tight coverage)
step_lat = small_radius_km / km_per_deg_lat * (math.sqrt(3) / 2)
step_lon = small_radius_km / km_per_deg_lon * 1.0
num_steps = int(large_radius_km / small_radius_km) + 2

# Function to extract coordinates from filename
def extract_coords(filename):
    match = re.search(r'center_([\d\.]+)_([\d\.]+)\.txt', filename)
    if match:
        return float(match.group(1)), float(match.group(2))
    return None

# Create map centered over Czech Republic
map_filtered_cover = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Store all coverage circle coordinates
cover_circle_coords = []

# Loop through each filtered center
for _, row in filtered.iterrows():
    center = extract_coords(row['source_file'])
    if not center:
        continue

    center_lat, center_lon = center

    # Draw original large circle
    folium.Circle(
        location=center,
        radius=large_radius_km * 1000,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.2,
        popup=f"Center: {center_lat:.6f}, {center_lon:.6f}"
    ).add_to(map_filtered_cover)

    # Generate coverage circles
    for i in range(-num_steps, num_steps + 1):
        for j in range(-num_steps, num_steps + 1):
            lat_offset = i * step_lat
            lon_offset = j * step_lon
            if i % 2 != 0:
                lon_offset += step_lon / 2
            new_lat = center_lat + lat_offset
            new_lon = center_lon + lon_offset
            dist_km = math.sqrt((lat_offset * km_per_deg_lat)**2 + (lon_offset * km_per_deg_lon)**2)
            if dist_km <= large_radius_km:
                cover_circle_coords.append((new_lat, new_lon))
                folium.Circle(
                    location=(new_lat, new_lon),
                    radius=small_radius_km * 1000,
                    color='blue',
                    fill=True,
                    fill_color='blue',
                    fill_opacity=0.3
                ).add_to(map_filtered_cover)

# Save and display the map
map_filtered_cover.save("filtered_circle_coverage_map.html")
map_filtered_cover
```

```python
# Combine coordinates and filenames into one list
cover_circle_data = []
data_dir = 'Cover center files'

for lat, lon in cover_circle_coords:
    filename = f"cover_center_{lat:.6f}_{lon:.6f}.txt"
    cover_circle_data.append({
        "center": (lat, lon),
        "filename": os.path.join(data_dir, filename)
    })

# Example usage
# for item in cover_circle_data:
    # print(f"Center: {item['center']}, Filename: {item['filename']}")
print(len(cover_circle_data))
cover_circle_data[:3]
```
