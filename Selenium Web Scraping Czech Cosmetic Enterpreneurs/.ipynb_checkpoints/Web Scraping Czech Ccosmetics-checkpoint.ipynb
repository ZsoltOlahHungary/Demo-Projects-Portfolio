```python
import folium

def shift_east(lat, lon, distance_km):
    km_per_degree_lon = 71
    delta_lon = distance_km / km_per_degree_lon
    return lat, lon + delta_lon

def shift_south(lat, lon, distance_km):
    km_per_degree_lat = 111
    delta_lat = distance_km / km_per_degree_lat
    return lat - delta_lat, lon

def add_circle(map_obj, center, radius_m, color, popup_text):
    folium.Circle(
        location=center,
        radius=radius_m,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.2,
        popup=popup_text
    ).add_to(map_obj)

# Parameters
initial_lat, initial_lon = 50.896450, 12.250195
circle_radius_m = 25000
distance_km = 50
rows = 6
cols = 10
circle_color = 'blue'

# Create map
map_czech = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Step 1: Store valid circle centers
circle_centers = []

for row in range(rows):
    row_lat, row_lon = shift_south(initial_lat, initial_lon, distance_km * row)
    for col in range(cols):
        row_index = row + 1  # 1-based indexing
        col_index = col + 1

        # Skip conditions
        if (row_index == 1 and col_index in [1, 2, 7, 8, 9, 10]) or \
           (row_index == 2 and col_index == 10) or \
           (row_index == 4 and col_index in [1]) or \
           (row_index == 5 and col_index in [1, 10]) or \
           (row_index == 6 and col_index in [1, 2, 9, 10]):
            continue

        lat, lon = shift_east(row_lat, row_lon, distance_km * col)
        circle_centers.append((lat, lon))

# Step 2: Draw circles from stored positions
for i, center in enumerate(circle_centers):
    add_circle(map_czech, center, circle_radius_m, circle_color, f'Circle {i+1}')

map_czech.save("czech_circles_map.html")

# Display map
map_czech
```

```python
# Step 3: Organize valid circle positions by grid index
circle_grid = {}
index = 0
for row in range(rows):
    row_lat, row_lon = shift_south(initial_lat, initial_lon, distance_km * row)
    for col in range(cols):
        row_index = row + 1
        col_index = col + 1

        if (row_index == 1 and col_index in [1, 8, 9, 10]) or \
           (row_index == 2 and col_index == 10) or \
           (row_index == 4 and col_index in [1]) or \
           (row_index == 5 and col_index in [1]) or \
           (row_index == 6 and col_index in [1, 2, 10]):
            continue

        lat, lon = shift_east(row_lat, row_lon, distance_km * col)
        circle_grid[(row, col)] = (lat, lon)

# Step 4A: Calculate and store gap centers
secondary_circle_centers = []

for row in range(rows - 1):
    for col in range(cols - 1):
        try:
            c1 = circle_grid[(row, col)]
            c2 = circle_grid[(row, col + 1)]
            c3 = circle_grid[(row + 1, col)]
            c4 = circle_grid[(row + 1, col + 1)]

            gap_lat = (c1[0] + c2[0] + c3[0] + c4[0]) / 4
            gap_lon = (c1[1] + c2[1] + c3[1] + c4[1]) / 4
            secondary_circle_centers.append((gap_lat, gap_lon))
        except KeyError:
            continue  # Skip if any of the four corners are missing

# Step 4B: Add gap circles to the map
for i, center in enumerate(secondary_circle_centers):
    add_circle(map_czech, center, circle_radius_m, 'red', f'Gap Filler {i+1}')

map_czech
```

```python
import folium

# Combine both primary and secondary circle centers
all_circle_centers = circle_centers + secondary_circle_centers

# Create a new map centered over the Czech Republic
map_combined = folium.Map(location=[49.8175, 15.4730], zoom_start=7)

# Add all circles to the new map
for i, center in enumerate(all_circle_centers):
    add_circle(map_combined, center, circle_radius_m, 'green', f'Unified Circle {i+1}')

# Display the new map
map_combined
```

```python
# Combine coordinates and filenames into one list
circle_data = []

for lat, lon in all_circle_centers:
    filename = f"center_{lat:.6f}_{lon:.6f}.txt"
    circle_data.append({
        "center": (lat, lon),
        "filename": filename
    })

# Example usage
for item in circle_data:
    print(f"Center: {item['center']}, Filename: {item['filename']}")
```

```python
import os
import pandas as pd
from bs4 import BeautifulSoup

data_dir = '.'
records = []

for filename in os.listdir(data_dir):
    if filename.startswith('center') and filename.endswith('.txt'):
        with open(os.path.join(data_dir, filename), 'r', encoding='utf-8') as file:
            soup = BeautifulSoup(file.read(), 'html.parser')

            for card in soup.select('div.card.sheet'):
                record = {'source_file': filename}

                # Extract name
                name_tag = card.select_one('h5')
                record['name'] = name_tag.get_text(strip=True) if name_tag else None

                # Extract registered_office and distance
                address_dd = card.select_one("div[title='Corresponding entity address'] dd")
                if address_dd:
                    span = address_dd.find("span", title="Distance from the location")
                    distance = span.get_text(strip=True) if span else None
                    full_address = address_dd.get_text(strip=True)
                    if span:
                        full_address = full_address.replace(distance, '').strip()
                    record['registered_office'] = full_address
                    record['distance_from_the_location'] = distance
                else:
                    record['registered_office'] = None
                    record['distance_from_the_location'] = None

                # Initialize fields
                record['legal_type'] = None
                record['role'] = None
                record['RN'] = None

                # Build label-value map from all <dt>/<dd> pairs
                label_map = {}
                for dl in card.select('dl.compact.tree'):
                    dt_tags = dl.select('dt')
                    dd_tags = dl.select('dd')
                    for dt, dd in zip(dt_tags, dd_tags):
                        label = dt.get_text(strip=True)
                        value = dd.get_text(strip=True)
                        label_map[label] = value

                # Assign values from label_map
                record['legal_type'] = label_map.get('Legal form') or label_map.get('Legal type') or next(iter(label_map.values()), None)
                record['role'] = label_map.get('Role')
                record['RN'] = label_map.get('RN')

                records.append(record)

df = pd.DataFrame(records)
df.head()
```
